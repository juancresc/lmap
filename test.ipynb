```python
import pandas as pd
df_map = pd.read_csv('map.csv', sep=',', comment="#", )
df_map.columns = ['marker', 'chromosome','position']
chrs = {}
for k,v in df_map.iterrows():
    pos = str(round(v.position, 2))
    if v.chromosome in chrs and pos in chrs[v.chromosome]:
        print(v.chromosome, pos)
    chrs.setdefault(v.chromosome, []).append(pos)
```

```python
df_map = pd.read_csv('map.csv', sep=',', comment='#')
df_map.columns = ['marker', 'LG', 'cM']
df_merger = pd.read_csv('merged_.csv', sep=',', comment="#", header=None)
max_col = max(df_merger.columns)
df_res = pd.merge(df_merger,df_map, left_on=0, right_on='marker')
df_res.cM = df_res.cM.round(2)
df_res.drop('marker', axis=1, inplace=True)
cols = [0,1,2,3,'LG','cM'] + [l for l in range(4, max_col + 1)]
df_res = df_res[cols]
df_res.rename({0: 'marker',1:'ref_chromosome',2:'ref_position', 3:'marker_position'}, axis=1, inplace=True)
df_res.to_csv('table.csv',sep=',', index=None)
```
