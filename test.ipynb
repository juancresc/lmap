```python
import pandas as pd
df = pd.read_csv('data.csv', sep=',', comment="#")
df.head(5)
```

```python
df = df.transpose()

df.loc['new_chr'] = 1
frames = [df.loc[['Marker','new_chr']], df[5:]]
df = pd.concat(frames)
df.rename(index={'Marker':'id'},inplace=True)
df.rename(index={'new_chr':''},inplace=True)
df.to_csv('qtl.csv', header=None)
```
